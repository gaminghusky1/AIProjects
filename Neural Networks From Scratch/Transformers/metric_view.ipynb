```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics = pd.read_csv("TinychatModels/tinychat_metrics.csv", index_col=0)
metrics
```

```python
loss_fig, loss_ax = plt.subplots()
loss_ax.plot(metrics["loss"])
loss_ax.set_xlabel("Step")
loss_ax.set_ylabel("Loss")
loss_ax.set_title("Loss Over Time")

ema_fig, ema_ax = plt.subplots()
ema_ax.plot(metrics["ema_loss"])
ema_ax.set_xlabel("Step")
ema_ax.set_ylabel("EMA Loss")
ema_ax.set_title("EMA Loss Over Time")

acc_fig, acc_ax = plt.subplots()
acc_ax.plot(metrics["accuracy"])
acc_ax.set_xlabel("Step")
acc_ax.set_ylabel("Accuracy")
acc_ax.set_title("Accuracy Over Time")
plt.show()
```
